```python
import pandas as pd
```

```python
aapl_df = pd.read_csv('aapl.csv')
amzn_df = pd.read_csv('amzn.csv')
fb_df = pd.read_csv('fb.csv')
goog_df = pd.read_csv('goog.csv')
nflx_df = pd.read_csv('nflx.csv')
```

```python
faang_concat = pd.concat([aapl_df, amzn_df, fb_df, goog_df, nflx_df], axis=0)
faang_concat
```

```python
aapl_df['ticker']='AAPL'
amzn_df['ticker']='AMZN'
fb_df['ticker']='FB'
goog_df['ticker']='GOOG'
nflx_df['ticker']='NFLX'
```

```python
faang_concat
```

```python
faang_concat.loc[:,'date'] = pd.to_datetime(faang_concat.date)
faang_concat.dtypes
```

```python
faang_concat
```

```python
faang_concat.sort_values(by=['date', 'ticker'], ascending=False)
faang_concat
```

```python
top_seven = faang_concat.nlargest(n=7, columns='volume')
top_seven
```

```python
faang_melt = faang_concat.melt(
 id_vars=['date', 'ticker'],
 value_vars=['open', 'high', 'low', 'close'],
 value_name='OHLC',
 var_name='Market_action'
)
faang_melt.to_csv('faang.csv', index = False)
faang_melt
```
